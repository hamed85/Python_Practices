# **A Practical Tutorial on Regular Expressions (Regex) in Python**

Regular Expressions (Regex) are patterns used to search, match, and manipulate text. Python provides built-in regex support through the re module.


1. Getting Started

```python
# import Regular Expressions
import re
```

A regex pattern is simply a string that describes what you're looking for.

```python
# Example
text = "My phone number is 555-1234"
match = re.search(r"\d+", text)
print(match.group())
```

Where:

"\d" = any digit (0-9)

"+" = one or more occurrences

r"" = raw string (recommended for regex patterns)

# **2. Common Regex Metacharacters**

| Pattern | Meaning                      |
| ------- | ---------------------------- |
| `.`     | Any character except newline |
| `\d`    | Digit                        |
| `\D`    | Non-digit                    |
| `\w`    | Letter, digit, underscore    |
| `\W`    | Non-word character           |
| `\s`    | Whitespace                   |
| `\S`    | Non-whitespace               |
| `^`     | Start of string              |
| `$`     | End of string                |


```python
# Example
import re
text = "abc123"
print(re.search(r"\d+", text).group())
```

# **3. Character Classes**
Character classes match one character from a set.

```python
# Match vowels
print(re.findall(r"[aeiou]", "hello world"))

# Match digits
print(re.findall(r"[0-9]", "abc123"))

# Negation
print(re.findall(r"[^0-9]", "abc123"))
```

# **4. Quantifiers**

Quantifiers specify how many times something should occur.

| Quantifier | Meaning         |
| ---------- | --------------- |
| `*`        | 0 or more       |
| `+`        | 1 or more       |
| `?`        | 0 or 1          |
| `{n}`      | Exactly n       |
| `{n,m}`    | Between n and m |

```python
# Example
re.findall(r"\d{3}", "123 45 6789") # The number of occurrences of digits is 3, meaning it matches 3 consecutive digits.
```

# **5. Searching vs Matching**
re.search(): it finds the first match anywhere:

```python
# Find the first match
re.search(r"cat", "The cat sleeps")
```

re.match(): it checks only the beginning:

```python
re.match(r"cat", "cat sleeps")
```

re.fullmatch(): entire string must match.

```python
# Example: returns a match
print(re.fullmatch(r"\d+", "12345"))

# Example: returns None
print(re.fullmatch(r"\d+", "123abc"))
```

# **6. Finding All Matches**
Use findall().

```python
# Example
text = "Alice has 3 cats and 2 dogs"
numbers = re.findall(r"\d+", text)
print(numbers)
```

# **7. Splitting Text**
Regex can split strings.

```python
text = "apple,banana;orange grape"
result = re.split(r"[,; ]+", text)
print(result)
```

# **8. Replacing Text**
Use sub().

```python
# Example
text = "Price: $100"
result = re.sub(r"\d+", "XXX", text)
print(result)
```

# **9. Groups and Capturing**
Parentheses create groups.

```python
# Example
text = "John Smith"
match = re.search(r"(\w+)\s+(\w+)", text)
print(match.group(1))
print(match.group(2))
```

```python
# Example for Named Groups:
text = "2025-12-31"
match = re.search(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", text)
print(match.group("year"))
```

# **10. Alternation (OR)**
Use | for alternation:

```python
re.findall(r"cat|dog", "cat dog bird")
```

# **11. Useful Real-World Examples**

```python
# Email Validation
pattern = r"^[\w.-]+@[\w.-]+\.\w+$"
email = "user@example.com"
if re.fullmatch(pattern, email):
    print("Valid")
```

```python
# Extract URLs
text = """
Visit https://google.com
or https://openai.com
"""

urls = re.findall(r"https?://\S+", text)
print(urls)
```

```python
# Extract Dates
text = "Meeting on 2025-08-15"
date = re.search(r"\d{4}-\d{2}-\d{2}", text)
print(date.group())
```

```python
# Parse Log Files:
log = "[ERROR] Database connection failed"
match = re.search(r"\[(.*?)\]", log)
print(match.group(1))
```

```python
# Compiling Patterns
import re
email_pattern = re.compile(r"^[\w.-]+@[\w.-]+\.\w+$")
print(email_pattern.fullmatch("test@example.com"))
```

```python
# Ignore Case
re.findall(r"python", "Python PYTHON", re.IGNORECASE)
```

```python
# Multiline
text = """first
second
third"""

re.findall(r"^.+", text, re.MULTILINE)
```

```python
# Cheat Sheet
'''
\d      digit
\w      word character
\s      whitespace

+       one or more
*       zero or more
?       optional

^       start
$       end

[]      character class
[^]     negated class

()      capture group
|       OR

{n}     exactly n
{n,m}   between n and m
'''
```

```python
import nltk
import regex as re

# re.search(pattern, string) looks for the first occurrence of a pattern anywhere in the string.
# If it finds a match, it returns a Match object; otherwise, it returns None.

# Example 1: Finding a word
text = "Groundhog is another name for woodchuck!"
result = re.search(r"[gG]roundhog|[Ww]oodchuck", text)

if result:
    print("Match found:", result.group())
    print("Starts at index:", result.start())
    print("Ends at index:", result.end())
```

```python
# Example 2: Searching for a number
# \d means "a digit (0–9)"
# + means "one or more occurrences"

text = "My order number is 12345."
result = re.search(r"\d+", text)

if result:
    print("Number found:", result.group())
```

```python
# Difference between search() and findall()
# search() → returns the first match only.
# findall() → returns all matches as a list.

text = "Groundhog is another name for woodchuck!"
print(re.search(r"[gG]roundhog|[Ww]oodchuck", text).group())
print(re.findall(r"[gG]roundhog|[Ww]oodchuck", text))
```

```python
text = 'That U.S.A. poster-print costs $12.40...'
pattern = r'''(?x)     # set flag to allow verbose regexps
(?:[A-Z]\.)+           # abbreviations, e.g. U.S.A.
| \w+(?:-\w+)*         # words with optional internal hyphens
| \$?\d+(?:\.\d+)?%?   # currency, percentages, e.g. $12.40, 82%
| \.\.\.               # ellipsis
| [][.,;"'?():_`-]     # these are separate tokens; includes ], [
... '''
nltk.regexp_tokenize(text, pattern)
```
